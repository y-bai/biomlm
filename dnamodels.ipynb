```python
LOCAL_MODEL_PATH_NT = 'pretrained/nucleotide-transformer-2.5b-1000g/'
```

```python
# load model from local disk

tokenizer = AutoTokenizer.from_pretrained(LOCAL_MODEL_PATH_NT, local_files_only=True)
# model = AutoModel.from_pretrained(LOCAL_MODEL_PATH_NT, local_files_only=True)
model = AutoModelForMaskedLM.from_pretrained(LOCAL_MODEL_PATH_NT, local_files_only=True)
```

```python
# Choose the length to which the input sequences are padded. By default, the 
# model max length is chosen, but feel free to decrease it as the time taken to 
# obtain the embeddings increases significantly with it.
max_length = tokenizer.model_max_length
# 1000
```

```python
# Create a dummy dna sequence and tokenize it
sequences = ["ATTCCGATTNNGATTCCG", "ATTTCTCTCTCTCTCTGAGATCGATCGATCGAT"]
tokens_ids = tokenizer.batch_encode_plus(sequences, return_tensors="pt", padding="max_length", max_length = max_length)["input_ids"]
```
